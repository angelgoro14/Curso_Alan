**Funciones, condicionales e iteraciones**

Vamos a crear una función de Python y llamarla desde un bucle o ciclo.

```
def HolaMundoXY(x, y):
  if (x < 10):
    print("Hola Mundo, x es < 10")
  elif (x < 20):
    print("Hola Mundo, x es >= 10 pero < 20")
  else:
    print("Hola Mundo, x es >= 20")
  return x + y

for i in range(8, 25, 5):  # i=8, 13, 18, 23 (inicio, paro, paso)
  print("--- Ejecutando HolaMundoXY("+str(i) + "," + str(i) + ")")
  salida = HolaMundoXY(i,i)
  print("Resultado de HolaMundoXY: " + str(salida))
```

```
#Llamemos nuevamente a la función HolaMundoXY
print(HolaMundoXY(1,2))
```

Si deseas un ciclo que comienza en 0 hasta 2 (exclusivo), puede hacer cualquiera de lo siguiente.

```
print("Iterar en elementos. `range(2)` regresa la lista [0,1]")
for i in range(2):
  print(i)

print("Iterar en una lista")
for i in [0,1]:
  print(i)

print("Ciclo mientras (while)")
i = 0
while i < 2:
  print(i)
  i += 1
```

```
print("En Python puedes usar continue and break")
while True:
  print("Dentro del ciclo while")
  break
```

**Numpy y listas**

Python tiene listas integradas en el lenguaje.
Sin embargo, utilizaremos una biblioteca llamada numpy para esto.
Numpy brinda muchas funciones que son útiles en Aprendizaje de Máquina.

Para usar Numpy necesitamos una declaración de importación. Esta declaración hace que todo el paquete numpy esté disponible usando la sintaxis abreviada 'np'.

```
import numpy as np  # Habilitar numpy con el alias np.

# Crear un arreglo de numpy y agregar un elemento
a = np.array(["Hola", "Mundo"])
a = np.append(a, "!")

# Imprimir el arreglo
print("Arreglo actual: ",a)

print("Imprimiendo cada elemento por separado")
for i in a:
  print(i)

print("\nImprimiendo cada elemento junto con su índice correspondiente")
for i,e in enumerate(a):
  print("índice: {}, corresponde a: {}".format(i, e)) # Usando .format para imprimir
                                                      # el primer corchete {} se sustituye por la variable i
                                                      # el segundo corchete {} se sustituye por la variable e
```

```
# Podemos realizar operaciones matemáticas en arreglos de numpy
print("\nOperaciones matemáticas básicas en arreglos de numpy")

b = np.array([0,1,4,3,2])  # Arreglo de numpy

#Imprimir el máximo
print("Max: {}".format(np.max(b)))
#Imprimir el promedio
print("Average (promedio): {}".format(np.average(b)))
```

```
print("\nPuedes imprimir el tipo de datos de las variables")
print("Tipo de b: {}, tipo de b[0]: {}".format(type(b), type(b[0])))
```

```
print("\nCrear un areglo de 3X3 números aleatorios con numpy")
c = np.random.rand(3, 3)
print(c)
```

```
print("\nPuedes imprimir las dimensiones de un arreglo")
print("Shape de a: {}".format(a.shape))
print("Shape de b: {}".format(b.shape))
print("Shape de c: {}".format(c.shape))
print("...Python usa tanto [0,1,2] como (0,1,2) para especificar listas")
```
