```
# 2._ Preparación de los datos (pre procesamiento)
df = pd.read_csv('https://raw.githubusercontent.com/AlanCIO/Disruptora/master/arrhythmia.data',header=None)
X = df.iloc[:,:-1]
y = df.iloc[:,-1]

# Se buscan valores nulos
nulos = []
for col in df:
  suma = (df[col] == "?").sum()
  if suma != 0:
    lista = [col, suma]
    nulos.append(lista)
	
print('[Columna, Elemento Nulo]:', nulos)
```

Se propone eliminar la columna 13 debido a la gran cantidad de elementos nulos que presenta, conteniendo un 84% de elementos nulos.



```
# Eliminar columna 13
df = df.drop([13], axis=1)
```

Para las columnas restantes se sugiere rellenar los datos faltantes con el promedio de el atributo en cuestion


```
# Se crea una funcion que permita obtener el promedio de la columna y asignarlo a la columna respectiva
import statistics                            # Libreria para obtener el promedio
def promedio(col):
  z = []
  for i in range(0,452):                     # Recorrido en las filas del DataFrame
    try:
      z.append(float(df.iloc[i, col]))       # Guardar los valores numericos en una lista
    except:
      pass                                   # Si hay un valor nulo o '?' ignorarlo y seguir

  mean = statistics.mean(z)                  # Obtener el promedio de z

  df[col] = df[col].replace('?', mean)       # Remplazar los '?' por el promedio de la columna
  z.clear()

  return mean
```

```
# Para columna 10
promedio(10)

# Para columna 11
promedio(11)

# Para columna 12
promedio(12)

# Para columna 14
promedio(14)

# Verificar que ya no existan mas valores nulos o con '?'
nulos = []
for col in df:
  suma = (df[col] == "?").sum()
  if suma != 0:
    lista = [col, suma]
    nulos.append(lista)
	
print('[Columna, Elemento Nulo]:', nulos)
```

```
# Importar clasificador
from sklearn import svm

# Crear modelo  #Usar cross validation para sintonizar los parámetros
clf = svm.SVC(C = 100, kernel = "poly", degree = 1)

# Asignacion de parametros de entrada
X = df.iloc[:,:-1]   
y = df.iloc[:,-1]

# Dividir datos en entrenamiento y prueba (evaluación)
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.30, random_state=42)     # 70% Entrenamiento - 30% Prueba

# Entrenar el modelo
clf.fit(X_train, y_train)

# Evaluar el modelo
from sklearn.metrics import accuracy_score
y_predict = clf.predict(X_test)
exactitud = accuracy_score(y_test, y_predict)
print(exactitud)
```

```
# Usando Cross-Validation
from sklearn.model_selection import cross_val_score

clf = svm.SVC(C = 100, kernel = "poly", degree = 1)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.30, random_state=42)     # 70% Entrenamiento - 30% Prueba
clf.fit(X_train, y_train)
y_predict = clf.predict(X_test)

scores = cross_val_score(clf, X, y, cv=3, scoring='accuracy')
print('Precisión: ', accuracy_score(y_test, y_predict))
print('Promedio Validacion Cruzada: ', scores.mean())
```

# **Regresión: Predicción de diabetes**


```
# Importar el DataSet de Diabetes
import seaborn as sns
import pandas as pd
from sklearn.datasets import load_diabetes
u,v = load_diabetes(return_X_y=True)

# Saber el tamaño del Frame
print('Tamaño: ', u.shape)
```

```
# Crear DataFrame de los datos
df_1 = pd.DataFrame(u, columns=['Edad', 'Sexo', 'bmi', 'bp', 'tc', 'ld1', 'hd1', 'tch', 'ltg', 'glu'])
print(df_1)
```

```
# Crear DataFrame de las etiquetas
df_2 = pd.DataFrame(v, columns=['Enfermedad'])
print(df_2)
```

```
# Juntar los DataFrame
dframe = pd.merge(df_1, df_2, left_index = True, right_index = True)
dframe
```

```
# Visualización de los valores 
sns.displot(dframe['Enfermedad'])
```

No es necesario normalizar los datos o realizar una limpieza de los mismos debido a que sklearn ya los tiene normalizados y ademas no se encuentran datos nulos o diferentes a valores numericos reales

```
# Dividir los datos en Datos de entrenamiento y prueba
from sklearn.model_selection import train_test_split
X_train, X_val, y_train, y_val = train_test_split(u, v, test_size = 0.1, random_state = 1)
X_train.shape, X_val.shape, y_train.shape, y_val.shape
```

```
# Importar clasificador
from sklearn import svm

# Crear modelo 
clf = svm.SVR(kernel="rbf", C=1000, gamma=100, epsilon=0.1)

# Entrenar el modelo
clf.fit(X_train, y_train)

# Evaluar el modelo
print(clf.score(X_train, y_train))

y_predict_1 = clf.predict(X_val)
```

```
import matplotlib.pyplot as plt

plt. figure(figsize=(10,10))
plt.scatter(y_val, y_predict_1, c ='crimson')
plt.yscale('log')
plt.xscale('log')

p1 = max(max(y_predict_1), max(y_val))
p2 = min(min(y_predict_1), min(y_val))
plt.plot([p1, p2], [p1, p2], 'b-')
plt.xlabel('Valores actuales', fontsize=15)
plt.ylabel('Predictions', fontsize=15)
plt.axis("equal")
plt.show()
```
