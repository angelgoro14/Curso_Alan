# ***EJERCICIO PARA MongoDB***

```python
import json
from pymongo import MongoClient 
  
  
# Establecer una concexión
myclient = MongoClient() 
   
# Crear una base de datos 
db = myclient["BDiris"]
   
# Crear una colección
coleccion = db["iris"]
  
# Abrimos el archivo Json (Se tiene que cargar el archivo en la parte de documentos de Colab)
with open('iris.json') as file:
    file_data = json.load(file)

# Agregar los datos del documento "iris.json" a la coleccion "coleccion"
if isinstance(file_data, list):
    coleccion.insert_many(file_data)  
else:
    coleccion.insert_one(file_data)
```

```python
# Imprimir toda la informacion del archivo
for x in coleccion.find():
  print(x)
```

### ***1._*** Encontrar el valor mínimo de la lóngitud del sépalo

```python
# Encontrar el valor mínimo para la longitud del sépalo
res = coleccion.find(projection={"_id": False, "sepalLength":True}).sort("sepalLength").limit(1)

for x in res:
  print('El valor minimo de longitud de sepalo es: ', x)
```

### ***2._*** Encontrar el valor máximo del ancho del pétalo


```python
#Encontrar el valor máximo para el ancho del pétalo

# UNA FORMA
#res = Collection.find(projection={"_id":False,"petalWidth":True})
#res_sorted = res.sort("petalWidth",-1)
#resultado = res_sorted.limit(1)

# OTRA FORMA
resultado = coleccion.find(projection={"_id":False,"petalWidth":True}).sort("petalWidth",-1).limit(1)
for x in resultado:
  print('El valor maximo de ancho de petalo es: ', x)
```

### ***3._*** Encontrar el promedio del ancho del sépalo

```python
# Encontrar el promedio del ancho del sépalo
import numpy as np
res = coleccion.find(projection={"_id": False, "sepalWidth":True})

lista = []
for x in res:
  lista.append(x['sepalWidth'])

arr = np.array(lista)
print('El valor promedio de ancho de sepalo es: ', np.mean(arr))
```

### ***4._*** Encontrar la desviación estándar de la longitud del pétalo

```python
# Encontrar la desviación estándar de la longitud del pétalo
import numpy as np
res = coleccion.find(projection={"_id": False, "petalLength":True})

lista = []
for x in res:
  lista.append(x['petalLength'])

arr = np.array(lista)
print('La desviacion estandar de longitud de petalo es: ', np.std(arr))
```

### ***5._*** ¿Cuál es el rango (max-min) de todos los atributos (Ancho/longitud del pétalo/sépalo)?
