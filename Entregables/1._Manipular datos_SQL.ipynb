# ***EJERCICIO PARA SQLITE3***

```python
# Importar las librerias necesarias
import sqlite3

# Realizar la coneccion con la base de datos
conn = sqlite3.connect("iris.db")
cursor = conn.cursor()
```

```python
# Importo pandas que me permite convertir a un dataframe y da una mejor presentacion
import pandas as pd

# DATAFRAME PARA ARCHIVO
# Selecciono todas las columnas de la tabla "Observation"
query_1 = '''SELECT *    
FROM Observation;'''

tabla_track = cursor.execute(query_1).fetchall()
columnas_track = cursor.execute('PRAGMA table_info("Observation")').fetchall()
df_Iris = pd.DataFrame(tabla_track, columns=[i[1] for i in columnas_track])

df_Iris.head(150)
```

### ***1._*** Encontrar el valor mínimo de la lóngitud del sépalo

```python
query = '''SELECT MIN(sepal_length)
FROM Observation'''

res = cursor.execute(query).fetchall()
print('El valor minimo de longitud de sepalo es:', res[0][0])
```

### ***2._*** Encontrar el valor máximo del ancho del pétalo

```python
query = '''SELECT MAX(petal_width)
FROM Observation'''

res = cursor.execute(query).fetchall()
print('El valor maximo de ancho de petalo es:', res[0][0])
```

### ***3._*** Encontrar el promedio del ancho del sépalo

```python
# Importar libreria que permite obtener el promedio
import statistics

# Seleccionar la columna de interes
query = '''SELECT sepal_width
FROM Observation'''

promedio = []                               # Crear una lista para guardar los valores de la columna
res = cursor.execute(query).fetchall()
for i in res:                               # Guardar los valores de la columna en la lista
  promedio.append(i[0])

mean = statistics.mean(promedio)                # Obtener el valor promedio de la lista
print('El valor promedio de ancho de sepalo es:', mean)
```

### ***4._*** Encontrar la desviación estándar de la longitud del pétalo

```python
# Importar libreria que permite obtener el promedio
import statistics

# Seleccionar la columna de interes
query = '''SELECT petal_length
FROM Observation'''

desviacion = []                               # Crear una lista para guardar los valores de la columna
res = cursor.execute(query).fetchall()
for i in res:                                 # Guardar los valores de la columna en la lista
  desviacion.append(i[0])

st_dev = statistics.pstdev(desviacion)              # Obtener la desviacion estandar
print('La desviación estandar de la longitud de petalo es:', st_dev)
```

### ***5._*** ¿Cuál es el rango (max-min) de todos los atributos (Ancho/longitud del pétalo/sépalo)?

```python
# Querys para obtener los maximos y minimos

# SepalLength
query_1 = '''SELECT MIN(sepal_length)
FROM Observation'''
query_10 = '''SELECT MAX(sepal_length)
FROM Observation'''
res = cursor.execute(query_1).fetchall()
res_1 = cursor.execute(query_10).fetchall()
print('El rango del atributo de SepalLength es:', res[0][0], '-', res_1[0][0])

# SepalWidth
query_2 = '''SELECT MIN(sepal_width)
FROM Observation'''
query_20 = '''SELECT MAX(sepal_width)
FROM Observation'''
res = cursor.execute(query_2).fetchall()
res_1 = cursor.execute(query_20).fetchall()
print('El rango del atributo de SepalLength es:', res[0][0], '-', res_1[0][0])

# PetalLength
query_3 = '''SELECT MIN(petal_length)
FROM Observation'''
query_30 = '''SELECT MAX(petal_length)
FROM Observation'''
res = cursor.execute(query_3).fetchall()
res_1 = cursor.execute(query_30).fetchall()
print('El rango del atributo de SepalLength es:', res[0][0], '-', res_1[0][0])

# PetalWidth
query_4 = '''SELECT MIN(petal_width)
FROM Observation'''
query_40 = '''SELECT MAX(petal_width)
FROM Observation'''
res = cursor.execute(query_4).fetchall()
res_1 = cursor.execute(query_40).fetchall()
print('El rango del atributo de SepalLength es:', res[0][0], '-', res_1[0][0])
```
